```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_moons, make_blobs

%matplotlib inline
```

```python
# Spiral dataset
n = 20
theta = np.sqrt(np.random.rand(n)) * 2 * np.pi
r_a = 2 * theta + np.pi
r_b = -2 * theta - np.pi
data_a = np.array([np.cos(theta) * r_a, np.sin(theta) * r_a]).T
data_b = np.array([np.cos(theta) * r_b, np.sin(theta) * r_b]).T
spiral_dataset = np.vstack((data_a, data_b))
spiral_labels = np.hstack((np.zeros(n), np.ones(n)))
spiral_df = pd.DataFrame(np.column_stack((spiral_dataset, spiral_labels)), columns=['x1', 'x2', 'y'])

# Moons dataset
moons_dataset, moons_labels = make_moons(n_samples=n*2, noise=0.1)
moons_df = pd.DataFrame(np.column_stack((moons_dataset, moons_labels)), columns=['x1', 'x2', 'y'])

# Two clusters dataset
two_clusters_dataset, two_clusters_labels = make_blobs(n_samples=n*2, centers=2, cluster_std=1.5)
two_clusters_df = pd.DataFrame(np.column_stack((two_clusters_dataset, two_clusters_labels)), columns=['x1', 'x2', 'y'])

# Plot datasets
datasets = [(spiral_dataset, spiral_labels, "Spiral dataset"),
            (moons_dataset, moons_labels, "Moons dataset"),
            (two_clusters_dataset, two_clusters_labels, "Two clusters dataset")]

for i, (data, labels, title) in enumerate(datasets):
    plt.subplot(1, 3, i+1)
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis")
    plt.title(title)
    plt.xticks([])
    plt.yticks([])


```

```python
moons_df.to_json('moons.json', orient='records')
two_clusters_df.to_json('two_clusters.json', orient='records')
```
