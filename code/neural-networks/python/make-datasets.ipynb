This notebook is used to make any underlying datasets required for the Neural Network MLU-Explain article.

Note, this is only for static datasets. Some of the data for the interactive charts is generated on the fly in the browser.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def step(x):
    return np.where(x < 0, -1, 1)

# activation dict
activations = {'sigmoid': sigmoid, 'tanh': tanh, 'relu': relu, 'step': step}

def plot_activation(name, xmin=-4, xmax=4, ymin=-1, ymax=1, save=True):
    """Plot and save data for activation function."""
    # Create an array of 30 evenly spaced values between -1 and 1
    x = np.linspace(xmin, xmax, 30)

    # Compute the tanh function for each value in x
    y = np.round(activations[name](x), 3)

    # Plot the function
    plt.plot(x, y)
    plt.ylim(ymin, ymax)
    
    if save:
        print(f"Saving data to {name}.json")
        # Create a dataframe from x and y
        data = {'x': x, 'y': y}
        df = pd.DataFrame(data)

        # Save the dataframe to json
        df.to_json(f"{name}.json", orient='records')

    # Add labels and show the plot
    plt.xlabel("x")
    plt.ylabel(f"{name}(x)")
    plt.show()
```

# sigmoid

```python
plot_activation('sigmoid')
```

# tanh

```python
plot_activation('tanh')
```

# ReLu

```python
plot_activation('relu')
```

# Step

```python
plot_activation('step', ymin=-5, ymax=4)
```
